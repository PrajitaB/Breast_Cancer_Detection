```python
# Breast Cancer Detection Using MIAS Dataset

import os
import numpy as np
import pandas as pd
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import lightgbm as lgb
from sklearn.neural_network import MLPClassifier
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import Callback
import seaborn as sns
import kagglehub

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Custom callback to print epoch details
class EpochLogger(Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        print(f"Epoch {epoch + 1}: loss = {logs.get('loss'):.4f}, accuracy = {logs.get('accuracy'):.4f}, "
              f"val_loss = {logs.get('val_loss'):.4f}, val_accuracy = {logs.get('val_accuracy'):.4f}")

# 1. Load and Preprocess the Dataset
def load_mias_data(dataset_path):
    info_path = os.path.join(dataset_path, 'all-mias', 'Info.txt')
    with open(info_path, 'r') as f:
        lines = f.readlines()

    data = []
    for line in lines[1:]:
        if line.strip():
            parts = line.strip().split()
            if len(parts) >= 4:
                refnum, _, _, severity = parts[:4]
                if severity in ['B', 'M']:
                    data.append({'refnum': refnum, 'label': severity})

    images = []
    labels = []
    image_dir = os.path.join(dataset_path, 'all-mias')
    for item in data:
        img_path = os.path.join(image_dir, f"{item['refnum']}.pgm")
        if os.path.exists(img_path):
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
                labels.append(item['label'])

    return images, labels

# 2. Image Preprocessing Function with Augmentation
def preprocess_image(img, target_size=(224, 224)):
    img = cv2.resize(img, target_size)
    img = cv2.equalizeHist(img)  # Histogram equalization
    img = cv2.GaussianBlur(img, (5, 5), 0)  # Noise reduction
    img = img / 255.0  # Normalization
    img_3ch = np.stack((img,) * 3, axis=-1)  # Convert to 3 channels
    return img, img_3ch

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 3. Display Processed Images
def display_processed_images(images, labels, num_images=5):
    plt.figure(figsize=(15, 5))
    for i in range(min(num_images, len(images))):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(f'Label: {labels[i]}')
        plt.axis('off')
    plt.show()

# 4. Load and Prepare Data
dataset_path = kagglehub.dataset_download("kmader/mias-mammography")
print("Path to dataset files:", dataset_path)
images, labels = load_mias_data(dataset_path)

# Preprocess and augment images
processed_images = []
processed_images_3ch = []
for img in images:
    img_proc, img_proc_3ch = preprocess_image(img)
    processed_images.append(img_proc)
    processed_images_3ch.append(img_proc_3ch)
    # Augment each image
    img_3ch_expanded = img_proc_3ch[np.newaxis, ...]
    for _ in range(2):  # Generate 2 augmented versions per image
        aug_iter = datagen.flow(img_3ch_expanded, batch_size=1)
        aug_img = next(aug_iter)[0]
        processed_images_3ch.append(aug_img)
        processed_images.append(aug_img[:, :, 0])  # Grayscale for traditional ML
        labels.append(labels[-1])  # Duplicate label for augmented image

# Display some processed images
display_processed_images(processed_images, labels)

# Convert to numpy arrays
X_images = np.array(processed_images_3ch)  # For deep learning
X_images_flat = np.array(processed_images).reshape(len(processed_images), -1)  # For traditional ML
le = LabelEncoder()
y = le.fit_transform(labels)  # B=0, M=1
num_classes = 2

# Split data
X_train_img, X_test_img, y_train, y_test = train_test_split(X_images, y, test_size=0.2, random_state=42)
X_train_flat, X_test_flat, _, _ = train_test_split(X_images_flat, y, test_size=0.2, random_state=42)

# 5. Define and Train Models
models = {}
y_pred_all = {}
y_prob_all = {}

def print_model_results(model_name, y_test, y_pred):
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred, target_names=['Benign', 'Malignant']))
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# 5.1 MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
model_mobilenet = Sequential([
    base_model,
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])
model_mobilenet.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("\nTraining MobileNetV2...")
model_mobilenet.fit(X_train_img, y_train, epochs=10, batch_size=32, validation_data=(X_test_img, y_test), callbacks=[EpochLogger()])
y_pred = np.argmax(model_mobilenet.predict(X_test_img, batch_size=32), axis=1)
y_prob = model_mobilenet.predict(X_test_img, batch_size=32)[:, 1]
models['MobileNetV2'] = model_mobilenet
y_pred_all['MobileNetV2'] = y_pred
y_prob_all['MobileNetV2'] = y_prob
print_model_results('MobileNetV2', y_test, y_pred)

# 5.2 CNN
model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("\nTraining CNN...")
model_cnn.fit(X_train_img, y_train, epochs=10, batch_size=32, validation_data=(X_test_img, y_test), callbacks=[EpochLogger()])
y_pred = np.argmax(model_cnn.predict(X_test_img, batch_size=32), axis=1)
y_prob = model_cnn.predict(X_test_img, batch_size=32)[:, 1]
models['CNN'] = model_cnn
y_pred_all['CNN'] = y_pred
y_prob_all['CNN'] = y_prob
print_model_results('CNN', y_test, y_pred)

# 5.3 ANN
model_ann = Sequential([
    Flatten(input_shape=(224, 224, 3)),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])
model_ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("\nTraining ANN...")
model_ann.fit(X_train_img, y_train, epochs=10, batch_size=32, validation_data=(X_test_img, y_test), callbacks=[EpochLogger()])
y_pred = np.argmax(model_ann.predict(X_test_img, batch_size=32), axis=1)
y_prob = model_ann.predict(X_test_img, batch_size=32)[:, 1]
models['ANN'] = model_ann
y_pred_all['ANN'] = y_pred
y_prob_all['ANN'] = y_prob
print_model_results('ANN', y_test, y_pred)

# 5.4 Traditional ML Models (10 classifiers)
traditional_models = {
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'SVM': SVC(probability=True, random_state=42),  # To be optimized later
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Naive Bayes': GaussianNB(),
    'LightGBM': lgb.LGBMClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'MLP': MLPClassifier(max_iter=1000, random_state=42)
}

# Optimize SVM with GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['rbf', 'linear'],
    'gamma': ['scale', 'auto', 0.1, 0.01]
}
svm = SVC(probability=True, random_state=42)
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_flat, y_train)
traditional_models['SVM'] = grid_search.best_estimator_
print(f"Best SVM parameters: {grid_search.best_params_}")

# Train all traditional models
for name, model in traditional_models.items():
    if name != 'SVM':  # SVM already trained via GridSearchCV
        print(f"\nTraining {name}...")
        model.fit(X_train_flat, y_train)
    y_pred = model.predict(X_test_flat)
    y_prob = model.predict_proba(X_test_flat)[:, 1]
    models[name] = model
    y_pred_all[name] = y_pred
    y_prob_all[name] = y_prob
    print_model_results(name, y_test, y_pred)

# 6. Plot ROC Curves
plt.figure(figsize=(12, 10))
for model_name in models.keys():
    fpr, tpr, _ = roc_curve(y_test, y_prob_all[model_name])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Models')
plt.legend(loc="lower right")
plt.show()
```

```python
# Breast Cancer Detection using MIAS Dataset
# Classifier: SVM, Hybridization: CNN, XGBoost, LightGBM
# Import all the libraries
import os
import numpy as np
import pandas as pd
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.svm import SVC
import xgboost as xgb
import lightgbm as lgb
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import Callback
import kagglehub

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Custom callback to print epoch details
class EpochLogger(Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        print(f"Epoch {epoch + 1}: loss = {logs.get('loss'):.4f}, accuracy = {logs.get('accuracy'):.4f}, "
              f"val_loss = {logs.get('val_loss'):.4f}, val_accuracy = {logs.get('val_accuracy'):.4f}")

# 1. Load and Preprocess the Dataset
def load_mias_data(dataset_path):
    info_path = os.path.join(dataset_path, 'all-mias', 'Info.txt')
    with open(info_path, 'r') as f:
        lines = f.readlines()

    data = []
    for line in lines[1:]:
        if line.strip():
            parts = line.strip().split()
            if len(parts) >= 4:
                refnum, _, _, severity = parts[:4]
                if severity in ['B', 'M']:
                    data.append({'refnum': refnum, 'label': severity})

    images = []
    labels = []
    image_dir = os.path.join(dataset_path, 'all-mias')
    for item in data:
        img_path = os.path.join(image_dir, f"{item['refnum']}.pgm")
        if os.path.exists(img_path):
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
                labels.append(item['label'])

    return images, labels

# 2. Image Preprocessing Function with Augmentation
def preprocess_image(img, target_size=(224, 224)):
    img = cv2.resize(img, target_size)
    img = cv2.equalizeHist(img)  # Histogram equalization
    img = cv2.GaussianBlur(img, (5, 5), 0)  # Noise reduction
    img = img / 255.0  # Normalization
    img_3ch = np.stack((img,) * 3, axis=-1)  # Convert to 3 channels
    return img, img_3ch

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 3. Display Processed Images
def display_processed_images(images, labels, title, num_images=5):
    plt.figure(figsize=(15, 5))
    for i in range(min(num_images, len(images))):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(f'{title} - Label: {labels[i]}')
        plt.axis('off')
    plt.show()

# 4. Load and Prepare Data
dataset_path = kagglehub.dataset_download("kmader/mias-mammography")
print("Path to dataset files:", dataset_path)
images, labels = load_mias_data(dataset_path)

# Preprocess and augment images
processed_images = []
processed_images_3ch = []
for img in images:
    img_proc, img_proc_3ch = preprocess_image(img)
    processed_images.append(img_proc)
    processed_images_3ch.append(img_proc_3ch)
    # Augment each image
    img_3ch_expanded = img_proc_3ch[np.newaxis, ...]
    for _ in range(2):  # Generate 2 augmented versions per image
        aug_iter = datagen.flow(img_3ch_expanded, batch_size=1)
        aug_img = next(aug_iter)[0]
        processed_images_3ch.append(aug_img)
        processed_images.append(aug_img[:, :, 0])  # Grayscale for traditional ML
        labels.append(labels[-1])  # Duplicate label for augmented image

# Convert to numpy arrays
X_images = np.array(processed_images_3ch)  # For deep learning
X_images_flat = np.array(processed_images).reshape(len(processed_images), -1)  # For traditional ML
le = LabelEncoder()
y = le.fit_transform(labels)  # B=0, M=1

# Split data
X_train_img, X_test_img, y_train, y_test = train_test_split(X_images, y, test_size=0.2, random_state=42)
X_train_flat, X_test_flat, _, _ = train_test_split(X_images_flat, y, test_size=0.2, random_state=42)

# Display processed images for all models (excluding AdaBoost)
display_processed_images(processed_images, labels, "CNN Processed Image")
display_processed_images(processed_images, labels, "MobileNetV2 Processed Image")
display_processed_images(processed_images, labels, "XGBoost Processed Image")
display_processed_images(processed_images, labels, "LightGBM Processed Image")

# 5. Feature Extraction Models
# 5.1 CNN Feature Extractor
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5)
])

cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("\nTraining CNN Feature Extractor...")
cnn_model.fit(X_train_img, y_train, epochs=10, batch_size=32, validation_data=(X_test_img, y_test), callbacks=[EpochLogger()])
X_train_cnn = cnn_model.predict(X_train_img)
X_test_cnn = cnn_model.predict(X_test_img)

# 5.2 MobileNetV2 Feature Extractor
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
mobilenet_model = Sequential([
    base_model,
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5)
])

mobilenet_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("\nTraining MobileNetV2 Feature Extractor...")
mobilenet_model.fit(X_train_img, y_train, epochs=10, batch_size=32, validation_data=(X_test_img, y_test), callbacks=[EpochLogger()])
X_train_mobilenet = mobilenet_model.predict(X_train_img)
X_test_mobilenet = mobilenet_model.predict(X_test_img)

# 5.3 XGBoost Feature Extractor
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
print("\nTraining XGBoost...")
xgb_model.fit(X_train_flat, y_train)
X_train_xgb = xgb_model.predict_proba(X_train_flat)
X_test_xgb = xgb_model.predict_proba(X_test_flat)

# 5.4 LightGBM Feature Extractor
lgbm_model = lgb.LGBMClassifier(random_state=42)
print("\nTraining LightGBM...")
lgbm_model.fit(X_train_flat, y_train)
X_train_lgbm = lgbm_model.predict_proba(X_train_flat)
X_test_lgbm = lgbm_model.predict_proba(X_test_flat)

# 6. SVM Classifiers
svm_cnn = SVC(kernel='rbf', probability=True, random_state=42)
svm_mobilenet = SVC(kernel='rbf', probability=True, random_state=42)
svm_xgb = SVC(kernel='rbf', probability=True, random_state=42)
svm_lgbm = SVC(kernel='rbf', probability=True, random_state=42)

# Train SVMs
print("\nTraining SVM with CNN features...")
svm_cnn.fit(X_train_cnn, y_train)
print("\nTraining SVM with MobileNetV2 features...")
svm_mobilenet.fit(X_train_mobilenet, y_train)
print("\nTraining SVM with XGBoost features...")
svm_xgb.fit(X_train_xgb, y_train)
print("\nTraining SVM with LightGBM features...")
svm_lgbm.fit(X_train_lgbm, y_train)

# Predictions and Probabilities
y_pred_cnn = svm_cnn.predict(X_test_cnn)
y_prob_cnn = svm_cnn.predict_proba(X_test_cnn)[:, 1]
y_pred_mobilenet = svm_mobilenet.predict(X_test_mobilenet)
y_prob_mobilenet = svm_mobilenet.predict_proba(X_test_mobilenet)[:, 1]
y_pred_xgb = svm_xgb.predict(X_test_xgb)
y_prob_xgb = svm_xgb.predict_proba(X_test_xgb)[:, 1]
y_pred_lgbm = svm_lgbm.predict(X_test_lgbm)
y_prob_lgbm = svm_lgbm.predict_proba(X_test_lgbm)[:, 1]

# 7. Evaluation
def print_evaluation(name, y_test, y_pred, y_prob):
    print(f"\n{name} Classification Report:")
    print(classification_report(y_test, y_pred, target_names=['Benign', 'Malignant']))

    cm = confusion_matrix(y_test, y_pred)
    print(f"\n{name} Confusion Matrix:")
    print(f"TP: {cm[1,1]}, TN: {cm[0,0]}, FP: {cm[0,1]}, FN: {cm[1,0]}")

print_evaluation("SVM with CNN", y_test, y_pred_cnn, y_prob_cnn)
print_evaluation("SVM with MobileNetV2", y_test, y_pred_mobilenet, y_prob_mobilenet)
print_evaluation("SVM with XGBoost", y_test, y_pred_xgb, y_prob_xgb)
print_evaluation("SVM with LightGBM", y_test, y_pred_lgbm, y_prob_lgbm)

# 8. ROC Curves
plt.figure(figsize=(12, 10))
for name, y_prob in [
    ("SVM with CNN", y_prob_cnn),
    ("SVM with MobileNetV2", y_prob_mobilenet),
    ("SVM with XGBoost", y_prob_xgb),
    ("SVM with LightGBM", y_prob_lgbm)
]:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for SVM Implementations')
plt.legend(loc="lower right")
plt.show()
```
