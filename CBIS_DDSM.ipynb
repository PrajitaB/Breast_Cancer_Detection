```python
# Breast Cancer Detection using CBIS DDSM
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Input, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import kagglehub
from tqdm import tqdm
import warnings
from imblearn.over_sampling import SMOTE
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelBinarizer

warnings.filterwarnings('ignore')

# Download CBIS-DDSM dataset
print("Downloading the CBIS-DDSM dataset...")
path = kagglehub.dataset_download('awsaf49/cbis-ddsm-breast-cancer-image-dataset')
dataset_path = path
csv_dir = os.path.join(dataset_path, "csv")
image_dir = os.path.join(dataset_path, "jpeg")

# Load CSV files
dicom_info = pd.read_csv(os.path.join(csv_dir, "dicom_info.csv"))
mass_train = pd.read_csv(os.path.join(csv_dir, "mass_case_description_train_set.csv"))
mass_test = pd.read_csv(os.path.join(csv_dir, "mass_case_description_test_set.csv"))
calc_train = pd.read_csv(os.path.join(csv_dir, "calc_case_description_train_set.csv"))
calc_test = pd.read_csv(os.path.join(csv_dir, "calc_case_description_test_set.csv"))

# Map images to labels
dicom_info['series_id'] = dicom_info['image_path'].apply(lambda x: x.split('/')[-2])
mass_train['series_id'] = mass_train['image file path'].apply(lambda x: x.split('/')[-2])
mass_test['series_id'] = mass_test['image file path'].apply(lambda x: x.split('/')[-2])
calc_train['series_id'] = calc_train['image file path'].apply(lambda x: x.split('/')[-2])
calc_test['series_id'] = calc_test['image file path'].apply(lambda x: x.split('/')[-2])

# Combine data
all_data = pd.concat([mass_train, mass_test, calc_train, calc_test], ignore_index=True)
merged_data = pd.merge(all_data, dicom_info[['series_id', 'image_path']], on='series_id', how='inner')

# Load and preprocess images
image_size = (224, 224)
images = []
labels = []
pathology_to_label = {'normal': 0, 'benign': 1, 'malignant': 2}

print("Loading and preprocessing images...")
for idx, row in tqdm(merged_data.iterrows(), total=len(merged_data)):
    series_id = row['series_id']
    image_filename = row['image_path'].split('/')[-1]
    image_path = os.path.join(image_dir, series_id, image_filename)
    if os.path.exists(image_path):
        try:
            img = load_img(image_path, target_size=image_size, color_mode='rgb')
            img_array = img_to_array(img) / 255.0
            pathology = row['pathology'].lower()
            if pathology == 'benign':
                if 'assessment' in row and row['assessment'] in [1, 2]:
                    pathology = 'normal'
            if pathology in pathology_to_label:
                label = pathology_to_label[pathology]
                images.append(img_array)
                labels.append(label)
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")

images = np.array(images)
labels = np.array(labels)
print(f"Loaded {len(images)} images and {len(labels)} labels.")

# Display some processed images
plt.figure(figsize=(15, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(images[i])
    plt.title(f"Label: {list(pathology_to_label.keys())[labels[i]]}")
    plt.axis('off')
plt.show()

# Split data
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42, stratify=labels)

# Feature extraction with MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
feature_extractor = Model(inputs=base_model.input, outputs=GlobalAveragePooling2D()(base_model.output))
train_features = feature_extractor.predict(X_train)
test_features = feature_extractor.predict(X_test)

# SMOTE for balancing
smote = SMOTE(random_state=42)
train_features_resampled, y_train_resampled = smote.fit_resample(train_features, y_train)

# Define models
models = {
    'MobileNetV2': Model(inputs=base_model.input, outputs=Dense(3, activation='softmax')(Dense(128, activation='relu')(feature_extractor.output))),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'SVM': SVC(kernel='rbf', probability=True, class_weight='balanced', random_state=42),
    'CNN': Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(3, activation='softmax')
    ]),
    'ANN': Sequential([
        Flatten(input_shape=(224, 224, 3)),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(128, activation='relu'),
        Dense(3, activation='softmax')
    ]),
    'KNN': KNeighborsClassifier(n_neighbors=3),
    'Random Forest': RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42),
    'Decision Tree': DecisionTreeClassifier(class_weight='balanced', random_state=42),
    'Logistic Regression': LogisticRegression(multi_class='multinomial', class_weight='balanced', max_iter=1000, random_state=42),
    'Naive Bayes': GaussianNB()
}

# Train and evaluate models
results = {}
roc_data = {}
class_names = ['normal', 'benign', 'malignant']

for name, model in models.items():
    print(f"\nTraining {name}...")
    if name in ['MobileNetV2', 'CNN', 'ANN']:
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        if name == 'MobileNetV2':
            history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32, verbose=1)
        else:
            history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=32, verbose=1)
        y_pred_prob = model.predict(X_test)
        y_pred = np.argmax(y_pred_prob, axis=1)
    else:
        model.fit(train_features_resampled, y_train_resampled)
        y_pred_prob = model.predict_proba(test_features)
        y_pred = model.predict(test_features)

    # Classification report and confusion matrix
    print(f"{name} Classification Report:")
    print(classification_report(y_test, y_pred, target_names=class_names))
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f'{name} Confusion Matrix')
    plt.show()

    # ROC data
    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)
    for i in range(3):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        roc_data[f"{name}_{class_names[i]}"] = (fpr, tpr, roc_auc)

# Plot common ROC curve
plt.figure(figsize=(10, 8))
for key, (fpr, tpr, roc_auc) in roc_data.items():
    plt.plot(fpr, tpr, label=f'{key} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Models')
plt.legend(loc='lower right')
plt.show()
```
